## Distribution Analysis


#### PDFs and CDFs
For a given distribution PDF(Probability Distribution Function),I would like to **see** how the Probability Distribution Function is and it would be really cool to see how the Cumulative Probability Distribution looks.

```python
import seaborn as sns

def distribution_analysis(n_sample_list,
                         distr_function = np.random.normal,
                         distr_kwargs = {'loc' : 25,
                                         'scale' : 25 }):
    '''
    Input:
    n_sample (list): list containing the number of observations to be sampled from distribution function
    distr_function (function) : The function of distribution to be sampled
    distr_kwargs (dictionary) : arguments of distribution function given in distr_function
    Output:
    Histogram plots about the given distribution function and cumulative probability function
    changing the number of samples.
    '''

    for sample_list in n_sample_list:

        for samples in range(sample_list):
            my_list = distr_function(**distr_kwargs,
                                     size=samples)

        #Creates two subplots and unpacks the output array immediately
        f, (ax1, ax2) = plt.subplots(1, 2, sharey=False,figsize = (10,4))
        ax1.hist(my_list,bins = 'auto')
        ax1.set_title('{} - Prob. Distr. Func - {} samples'.format(distr_function.__name__.upper() , sample_list))
        ax2.hist(my_list,cumulative=True,density = True,bins = 'auto')
        ax2.set_title('{} - Cum. Distr Func - {} samples'.format(distr_function.__name__.upper(), sample_list))
        mean = np.mean(my_list)
        std = np.std(my_list)
        plt.show()
        print('Mean Value {}'.format(mean))
        print('Standard Deviation Value {}'.format(std))
        
distribution_analysis([1000])
```

```python
distribution_analysis([1000,5000,10000],np.random.beta,{'a' : 0.4,'b' : 0.2})
distribution_analysis([1000,5000,10000],np.random.beta,{'a' : 2,'b' : 1})
distribution_analysis([1000,5000,10000],np.random.beta,{'a' : 20,'b' : 10})
print()
distribution_analysis([1000,5000,10000],np.random.exponential,{'scale':0.5})
distribution_analysis([1000,5000,10000],np.random.exponential,{'scale':10})

distribution_analysis([1000,5000,10000],np.random.chisquare,{'df':4.5})
```

```python
def central_limit_theorem(distribution = np.random.uniform,
                          distribution_kwars = {'low':0,'high':1},
                          number_samples_list = [1,2,5,10,15,20,30,35,50],
                          POPULATION = 10000):
    
    #Define the population size
    distribution_kwars['size'] = POPULATION
    
    #Generate the array about our population
    population_array = distribution(**distribution_kwars)
    
    #Create a list(my_choice) that selects n observations for the sampling
    for n in number_samples_list:
        my_choice =  list()
        
        for i in range(POPULATION):
            sampling = random.choices(population_array,k = n)
            my_choice.append(sampling)
        
        #transform in dataframe for easier data manipulation
        exp_df = pd.DataFrame(my_choice)
        exp_df['mean_sample'] = exp_df.mean(axis=1)
        
        #Create our plots
        f,(ax1,ax2) = plt.subplots(1, 2, sharey=False,sharex = False, figsize = (12,6))

        ax1.hist(exp_df['mean_sample'],bins = 'auto',alpha = 0.7,density=True,color = 'red',label = 'Bootstrapping - {} samples'.format(n))
        ax1.hist(population_array,alpha = 0.6,bins = 'auto',density=True,color = 'blue',label = 'PDF')
        ax1.legend(loc = 2)
        ax1.grid(color = 'black')

        ax2.hist(exp_df['mean_sample'],bins = 'auto',alpha = 0.8,density=True,color = 'red',label = 'Bootstrapping - {} samples'.format(n))
        plt.grid(color = 'black')
        ax2.legend(loc = 2)
        
        #Print some stats
        print('Obtivemos como media do Bootstrapping : {}'.format(exp_df['mean_sample'].mean()))
        print('Obtivemos como desvio padrao do Bootstrapping : {}'.format(exp_df['mean_sample'].std()))
        print()
        print('Obtivemos como media do PDF : {}'.format(np.mean(population_array)))
        print('Obtivemos como desvio padrao do PDF : {}'.format(np.std(population_array)))    
        plt.show()
        
central_limit_theorem(np.random.exponential,
                     distribution_kwars = {'scale':5.0})
```

## Dice simulation

Cells for making a dataframe that creates all possibilities of a 6-faced dices rolls

```python
dado = [1, 2, 3, 4, 5, 6]

dice_list = [dado,dado,dado,dado,dado,dado]
index = pd.MultiIndex.from_product(dice_list, names = ['dado'+ str(i + 1) for i in range(len(dice_list))])

jogadas = pd.DataFrame(index = index).reset_index()
jogadas['soma'] = jogadas.sum(axis=1)


n = len(set(jogadas['soma']))
bins=np.arange(1,n+2*len(dice_list),1)

fig, ax = plt.subplots(1,1,figsize = (17,10))
ax.hist(jogadas['soma'], bins=bins, align='left',density = True)
ax.set_xticks(bins[:-1])
plt.grid(which= 'major')
plt.minorticks_on()
plt.show()
```

```python
freq_table = jogadas[['soma','dado1']].groupby('soma').count()
freq_table['probability (%)'] = (freq_table['dado1']/freq_table['dado1'].sum())*100
freq_table['accumulated'] = np.cumsum(freq_table['probability (%)'])
freq_table['gain (%)'] = freq_table.accumulated - freq_table.accumulated.shift(1)
```

For a 6 faced dice, if the numbers from 18 to 23 are bennefitial for us we have a 51.47 % winning rate, it means that if the weights of winning and losing are equal, we would bias for our winning.
There are 30 probabilities of sums for a 6 faced-dice and we just need 5 to win in long-runs :)

Let's imagine each round of the game we're playing we bets $ 1000 in a cassino... 

The gambler plays 6 dices and if the sum of the dices is between 18 and 23 we gain 1000 dollars, otherwise we lose 1000 dollars!

Imagine each round takes about 1 min and we're playing for 3 hours, it means we played 180 rounds.
With the help of python we can simulate this scenario 1000 times and we can have a estimation of gain

```python
from tqdm import tqdm

full_round_list = []

for full_round in tqdm(range(1000)):
    
    win_lose_list = list()
    for i in range(180):
        win_lose_list.append(random.choice(jogadas.soma))

    show_gains = [-1 if (number > 23) | (number < 18) else 1 for number in win_lose_list]
    full_round_list.append(np.sum(show_gains))
    
plt.hist(full_round_list)
plt.xlabel('Gains - Losses')
plt.ylabel('Occurances')
plt.grid()
plt.show()
print(np.mean(full_round_list))
```

We can see that the distribution of Gain - Losses looks like a Normal distribution, it happens because of the CLT(Central Limit Theorem)
